```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
x = np.linspace(-10, 10, num=N )
```

```python
def g(x):
    return x**2

def f(x):
    return np.cos(x)

y = f(g(x))
```

```python
plt.plot(x, y)
plt.grid()
```
