```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
df.groupby(["sex","time"])[["total_bill"]].mean()
```

```python
df_group = df.groupby(["sex","time"])[["total_bill"]].mean().reset_index()
df_group
```

```python
df_group.pivot_table(values="total_bill", index="sex", columns="time")
```

```python
df.pivot_table(values="total_bill", index="sex", columns="time")
```

```python
df.pivot_table(values="total_bill", index="sex", columns="time", aggfunc=(np.mean))
```

```python
df.pivot_table(values="total_bill", index="sex", columns="time", aggfunc=([np.mean, np.std, np.median, np.max,np.sum]))
```

```python
df_pivot = df.pivot_table(values="total_bill", index="sex", columns="time", aggfunc=([np.mean, np.std, np.median, np.max,np.sum]))
df_pivot
```

```python
df_pivot.unstack()
```

```python
df_pivot.unstack().reset_index()
```
